```python
import pandas as pd
```

```python
path="/content/drive/MyDrive/Colab Notebooks/smsdatanewdataset.csv"
```

```python
df=pd.read_csv(path)
```

```python
df
```

```python
df=df[['address','body','BankTrancation']]
```

```python
df
```

```python
low_body=[]

for data in df['body']:
  data=data.lower()
  low_body.append(data)
```

```python
df['body']=low_body
```

```python
df.head()
```

```python
low_banktr=[]
for tr in df['BankTrancation']:
  tr=tr.lower()
  low_banktr.append(tr)
```

```python
df['BankTrancation']=low_banktr
```

```python
df.head()
df.describe()
```

```python
df['BankTrancation'].value_counts()
```

```python
import string

def remove_punctuation(text):
  for punctuation in string.punctuation:
    text=text.replace(punctuation,'')
  return text
```

```python
df['body']=df['body'].apply(remove_punctuation)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vector=CountVectorizer()
```

```python
smsbody=list(df['body'])
type(smsbody)
print(smsbody)
```

```python
count_vector.fit(smsbody)
```

```python
word_list = count_vector.get_feature_names()
len(word_list)
```

```python
doc_array = count_vector.transform(smsbody).toarray()
doc_array
```

```python
frequency_matrix = pd.DataFrame(doc_array,
                                columns = word_list)
frequency_matrix
```

```python
text_counts=count_vector.fit_transform(df['body'])
```

```python
text_counts
```

```python
x=df['body']
y=df['BankTrancation']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=5,test_size=0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import naive_bayes, metrics
```

```python
mnb=MultinomialNB()
```

```python
# mnb.fit(x_train,y_train)
x_train_count=count_vector.transform(x_train)
x_test_count=count_vector.transform(x_test)
```

```python
classifier =naive_bayes.MultinomialNB()
classifier.fit(x_train_count, y_train)

predictions = classifier.predict(x_test_count)
```

```python
#create dataframe for y-test and y-pred result show
```

```python
predictions
```

```python
print(metrics.accuracy_score(predictions, y_test))
```
